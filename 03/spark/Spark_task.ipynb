```python
import pandas as pd
import random
```

```python
is_user = []
priority = []
subscribers = []
timestamp = []
```

```python
for i in range(1000):
    is_user.append(random.randint(0, 1))
    priority.append(random.randint(0,100))
    subscribers.append(random.randint(0,6000))
    timestamp.append(1600000000 + random.randint(1,10000000))
```

```python
data = {'is_user': is_user,
        'priority': priority,
        'subscribers_number': subscribers,
        'timestamp': timestamp
       }
df = pd.DataFrame(data)
```

```python
df.to_csv('spark_task.csv')
```

```python
import pyspark
sc = pyspark.SparkContext('local[*]')

```

```python
from pyspark import SparkContext
from pyspark.streaming import StreamingContext
```

```python
session = pyspark.sql.SparkSession(sc)
```

```python
table = session.read.csv('/spa_task/spark_task.csv', header = True)
```

```python
a = table.select("*").where("is_user = 0").orderBy("priority", ascending = False).limit(100)
```

```python
b = table.select("*").where("is_user = 1").orderBy("priority", ascending = False).limit(50)
```

```python
c = a.union(b)
```

```python
d = c.where("subscribers_number > 2000").orderBy("timestamp").union(c.where("subscribers_number <= 2000").orderBy("timestamp"))
```

```python
d.coalesce(1).write.format("csv").save("/spa_answer")
```
